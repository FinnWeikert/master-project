```python
import pandas as pd
```

```python
ratings = pd.read_csv('averaged_ratings.csv')
raw_scores = pd.read_csv('PerformanceAssessmentData_finish_corrected_filtered.csv', sep=';')
```

```python
ratings.head()
```

```python
raw_scores.head()
```

```python
"""
Case_Number <=> Case Type mapping:
1 - radius
2 - malleolus
3 - Ulna
"""
raw_scores['Case_Number'] = raw_scores['Case Type'].map({'radius': 1, 'malleolus': 2, 'Ulna': 3})
avg_raw_scores = raw_scores.groupby(['Participant Number', 'Case_Number'])['QRS Overal'].mean().reset_index()
df_merged = pd.merge(ratings, avg_raw_scores[['QRS Overal', 'Case_Number', 'Participant Number']], on=['Participant Number', 'Case_Number'], how='inner')
# drop rows with missing GRS_Total
df_merged = df_merged.dropna(subset=['QRS Overal'])
df_merged.rename(columns={'QRS Overal': 'QRS_Overal'}, inplace=True)
# save as csv
df_merged.to_csv('merged_scores.csv', index=False)
```
